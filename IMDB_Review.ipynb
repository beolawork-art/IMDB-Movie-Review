```python
import pandas as pd
import re
```

```python
df = pd.read_csv('IMDB Dataset.csv')
```

```python
# We make it lowercase, remove spaces, and turn it into 1s and 0s
df['sentiment'] = df['sentiment'].astype(str).str.strip().str.lower()
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})

# Create the Clean Review column
# This removes the <br /> tags and extra symbols
df['clean_review'] = df['review'].str.replace(r'<[^>]*>', ' ', regex=True) # Removes HTML tags
df['clean_review'] = df['clean_review'].str.replace(r'[^a-zA-Z\s]', '', regex=True).str.lower()

# The "Super Eraser" (Fixes the 0x81 Encoding Error)
def remove_weird_chars(text):
    return text.encode('ascii', 'ignore').decode('ascii')

df['review'] = df['review'].apply(remove_weird_chars)
df['clean_review'] = df['clean_review'].apply(remove_weird_chars)

# Save the Perfect File
df.to_csv('IMDB_Perfect_Clean.csv', index=False, encoding='utf-8')
print("✅ Success! Your 'IMDB_Perfect_Clean.csv' is ready. Download it now!")
```
